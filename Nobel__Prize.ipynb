## The Nobel Prize is one of the world's most well known scientific award that was established by Alfred Nobel.  Every year it's given to scientists and scholars in chemistry, literature, physics, phsiology/medicine, economics and peace.  The first Nobel Prize was handed out in 1901 and the Nobel foundation has made a dataset of all prize winners from 1901 to 2016.

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Reading in the Nobel Prize data
nobel = pd.read_csv('datasets/nobel.csv')
nobel.rename(columns={"sex": "gender"}, inplace = True)
# Taking a look at the first several winners
nobel.head(6)
```

## Looking at the first couple Nobel laureates is the individual who discovered x-rays, Wilhelm Conrad Rontgen. Explore which country has had the most nobel prizes from 1901 to 2016.

```python
# Display the number of (possibly shared) Nobel Prizes handed
# out between 1901 and 2016
display(len(nobel['prize']))

# Display the number of prizes won by male and female recipients.
display(nobel.gender.value_counts())

# Display the number of prizes won by the top 10 nationalities.
nobel.birth_country.value_counts().head(10)
```

## Explore the proportions of Nobel Prizes in the United States over the decades.

```python
# Calculating the proportion of USA born winners per decade
nobel['usa_born_winner'] = nobel['birth_country'] =='United States of America'
nobel['decade'] = (np.floor((nobel['year'])/10) * 10).astype(int)
prop_usa_winners = nobel.groupby('decade', as_index=False)['usa_born_winner'].mean()

# Display the proportions of USA born winners per decade
display(prop_usa_winners)
```

## Plot the proprtions of Nobel Prizes in the United States over the decades.

```python
# Setting the plotting theme
sns.set()
# and setting the size of all plots.
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [11, 7]

# Plotting USA born winners 
ax = sns.lineplot(x='decade',y='usa_born_winner', data=prop_usa_winners)

# Adding %-formatting to the y-axis
from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter(1.0))
```

## Data exploration shows a high proportion of Nobel Prizes in the United States starting in the 1930s.  Data analysis also shows 95% Nobel Prizes have been awarded to males and 5% to females.  Is there a difference in certain categories more than others?

```python
# Calculating the proportion of female laureates per decade
nobel['female_winner'] = nobel['gender'] == 'Female'
prop_female_winners = nobel.groupby(['decade','category'], as_index=False)['female_winner'].mean()

# Plotting USA born winners with % winners on the y-axis

ax = sns.lineplot(x='decade',y='female_winner',hue='category', data=prop_female_winners)
ax.yaxis.set_major_formatter(PercentFormatter(1.0))
```

## The plot above shows the imbalance is pretty large with physics, economics, and chemistry.  Medicine has a somewhat positive trend, and since the 1990s the literature prize is more balanced.  According to the plot, from 2010 to 2016 the peace prize has the highest percentage.

```python
# Picking out the first woman to win a Nobel Prize
nobel[nobel.gender == 'Female'].nsmallest(1,'year')
```

## Which Nobel Laureates have won a Nobel Prize more than once?

```python
# Selecting the laureates that have received 2 or more prizes.
nobel.groupby('full_name').filter(lambda group: len(group) > 1)
```

## Data exploration showed that Marie Curie got the prize in physics for discovering radiation and in chemistry for isolating radium and polonium and that John Bardeen got it twice in physics for transistors and superconductivity. Frederick Sanger got a Nobel prize twice in chemistry, and Linus Carl Pauling got it first in chemistry and later in peace for his work in promoting nuclear disarmament.  The organizations Red Cross and the UNHCR have also gotten the Nobel Prize twice.  Are there trends in age of Nobel Prize recipient?

```python
# Converting birth_date from String to datetime
nobel['birth_date'] = pd.to_datetime(nobel['birth_date'])

# Calculating the age of Nobel Prize winners
nobel['age'] = nobel['year'] - nobel['birth_date'].dt.year

# Plotting the age of Nobel Prize winners
sns.lmplot(x='year',y='age', data=nobel)
```

## The plot above shows that in the past people use to be around 55 when they received a prize, but nowadays the average is closer to 65. There is also a large spread in the laureates' ages, since some are very young while most are 50+.

## The plot shows the density of points is higher nowadays than in the early 1900s since there are more winners due to more prizes being shared. There was also a disruption in awarded prizes around the Second World War (1939 - 1945).   Are there age trends within the different categories?

```python
# Same plot as above, but separate plots for each type of Nobel Prize
sns.lmplot(x='year',y='age',row = 'category', data=nobel)
```

## The plots above show that the winners of chemistry, medicine, and physics prizes have gotten older over time. The trend is strongest for physics where the previous average prize age was 50 while now it's almost 70.  Literature and economics are more stable and economics appears to be a newer category. The peace prize shows the opposite trend where winners are getting younger.  In fact, who are the oldest and youngest people ever to have won a Nobel Prize?
 

```python
# The oldest winner of a Nobel Prize as of 2016
display(nobel.nlargest(1,'age'))

# The youngest winner of a Nobel Prize as of 2016
nobel.nsmallest(1,'age')
```
